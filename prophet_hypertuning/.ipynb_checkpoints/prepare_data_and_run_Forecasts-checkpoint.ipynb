```python
def return_read_query(metric:str,source_table:str) ->str:
    """Function that returns PW aggregation query prior to forecasting"""

    if metric =="playtime":
        return f"""
            SELECT  checkpoint,platform,vxp_client_id,game_id,
            sum(Total_session_length)/3600 as playtime FROM {source_table} 
            where vxp_client_id=@vxp_client_id and 
            checkpoint < @date_until
            group by checkpoint,platform,vxp_client_id,game_id
            order by checkpoint  
            """
    elif metric =="asl":
        return f"""
            SELECT  checkpoint,platform,vxp_client_id,game_id,
            sum(Total_session_length)/ nullif(sum (Session_count),0) as asl FROM {source_table} 
            where vxp_client_id=@vxp_client_id and 
            checkpoint < @date_until
            group by checkpoint,platform,vxp_client_id,game_id
            order by checkpoint  
            """
    elif metric=="dau":
        return f"""
            SELECT  checkpoint,platform,vxp_client_id,game_id,
            sum(dau) as dau FROM {source_table} 
            where vxp_client_id=@vxp_client_id and 
            checkpoint < @date_until
            group by checkpoint,platform,vxp_client_id,game_id
            order by checkpoint  
            """

def read_metric(date_until,vxp_client_id,bq_source_table,metric):
    query_string=return_read_query(metric,bq_source_table)    
    job_config = bigquery.QueryJobConfig(
    query_parameters=[
        bigquery.ScalarQueryParameter("vxp_client_id", "INT64", vxp_client_id),
        bigquery.ScalarQueryParameter("date_until", "STRING", date_until)
    ]
    )

    metric_df=(bqclient.query(query_string,job_config=job_config)
            .result()
            .to_dataframe(create_bqstorage_client=True,) ) 
    last_agg_date=metric_df.iloc[-1,0]

                
    return metric_df,last_agg_date
```
